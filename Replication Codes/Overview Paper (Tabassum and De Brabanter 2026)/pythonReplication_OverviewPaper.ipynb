```python
from recombinator.optimal_block_length import optimal_block_length
from arch.bootstrap import optimal_block_length
import pandas as pd
import numpy as np
import math
```

```python
data = np.loadtxt('simEx1.csv', delimiter=',')
nsims = 100
n = 200

index_rec = 0
bopt_cbb_rec = np.empty(shape=nsims, dtype=float)
bopt_sbb_rec = np.empty(shape=nsims, dtype=float)

for iter in range(nsims-1):
    y = data[index_rec:(index_rec+n-1)]
    b_star = optimal_block_length(y)
    bopt_sbb_rec[iter] = b_star[0].b_star_sb
    bopt_cbb_rec[iter] = math.ceil(b_star[0].b_star_cb)
    index_rec = index_rec+n 
```

```python
np.savetxt('sbb_recomb.csv', bopt_sbb_rec, delimiter=',')
np.savetxt('cbb_recomb.csv', bopt_cbb_rec, delimiter=',')
```

```python
data = np.loadtxt('simEx1.csv', delimiter=',')
nsims = 100
n = 200

index_rec = 0
bopt_cbb_arch = np.empty(shape=nsims, dtype=float)
bopt_sbb_arch = np.empty(shape=nsims, dtype=float)

for iter in range(nsims-1):
    y = data[index_rec:(index_rec+n-1)]
    b_star = optimal_block_length(y)
    bopt_sbb_arch[iter] = b_star.iloc[0,0]
    bopt_cbb_arch[iter] = math.ceil(b_star.iloc[0,1])
    index_rec = index_rec+n 
```

```python
np.savetxt('sbb_arch.csv', bopt_sbb_arch, delimiter=',')
np.savetxt('cbb_arch.csv', bopt_cbb_arch, delimiter=',')
```
